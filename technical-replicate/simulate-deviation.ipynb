```python
import scipy
import numpy as np
import matplotlib.pyplot as plt
```

```python
%matplotlib inline
```

# Random Start 

```python
x0 = np.random.normal(1,.5,size=10)
x0
```

```python
xsim = np.zeros((50,x0.shape[0]))
xsim[0,:] = x0

r = .1

for i in range(1,xsim.shape[0]):
    xsim[i,:] = xsim[i-1,:]*(1+r)
    xsim[i,:] = np.max(np.column_stack((xsim[i,:],1e-6+np.zeros(x0.shape[0]))),1)
```

```python
plt.figure(figsize=(12,4))
plt.subplot(131)
plt.plot(xsim);
plt.title("y(x)")

plt.subplot(132)
plt.plot(np.log2(xsim));
plt.title("$log$ y(x)")

plt.subplot(133)
plt.plot(np.log2(xsim)-np.log2(xsim).mean(1)[:,None]);
plt.title("$log$ y(x) - $\mu$")
```

# Random rate
